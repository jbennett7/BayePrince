# Key Terms and Concepts

__ad valorem tax__, A percentage tax, such as the sales tax. The tax is higher for high-priced items.

__change in demand__, Whenever advertising, income, or the price of related goods changes, it leads to a change in demand; the position of the entire demand curve shifts. A rightward shift in the demand curve is called an _increase in demand_, since more of the good is demanded at each price. A leftward shift in the demand curve is called a _decrease in demand_.

```r
curve(5-x,from=0,to=5,main="Changes in Demand",ylab="Price", xlab="Quantity",bty='n',axes=FALSE)
axis(side=1,labels=FALSE)
axis(side=2,labels=FALSE)
curve(6-x,from=0,to=5,add=TRUE,col='blue')
curve(4-x,from=0,to=5,add=TRUE,col='blue')
arrows(x0=2.6,y0=2.2,x1=2,y1=2.2,length=0.1,angle=30,col='blue',lwd=2)
text(x=.9,y=2.2,'Decrease in demand')
arrows(x0=2.4,y0=2.8,x1=3,y1=2.8,length=0.1,angle=30,col='blue',lwd=2)
text(x=4.1,y=2.8,'Increase in demand')
```

__change in quantity demanded__, When we graph the demand curve for good $X$, we hold everything but the price of $X$ constant. A representative demand curve is given by the center curve above. A movement along this curve is called a change in quantity demanded.

__change in quantity supplied__, A movement along the supply curve.

__change in supply__, Whenever one or more of the _supply shifter_ variables change the entire supply curve shifts.

```r
curve(x-0,from=0,to=5,main="Changes in Suppy",ylab="Price", xlab="Quantity",bty='n',axes=FALSE)
axis(side=1,labels=FALSE)
axis(side=2,labels=FALSE)
curve(x-1,from=0,to=5,add=TRUE,col='blue')
curve(x+1,from=0,to=5,add=TRUE,col='blue')
arrows(x0=2,y0=2.3,x1=1.5,y1=2.3,length=0.1,angle=30,col='blue',lwd=2)
text(x=.6,y=2.4,'Decrease in supply')
arrows(x0=2.6,y0=2.3,x1=3.2,y1=2.3,length=0.1,angle=30,col='blue',lwd=2)
text(x=4.2,y=2.3,'Increase in supply')
```
