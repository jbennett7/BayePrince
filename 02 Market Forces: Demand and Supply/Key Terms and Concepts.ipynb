# Key Terms and Concepts

__ad valorem tax__, A percentage tax, such as the sales tax. The tax is higher for high-priced items.

__change in demand__, Whenever advertising, income, or the price of related goods changes, it leads to a change in demand; the position of the entire demand curve shifts. A rightward shift in the demand curve is called an _increase in demand_, since more of the good is demanded at each price. A leftward shift in the demand curve is called a _decrease in demand_.

```r
curve(5-x,from=0,to=3.5,main="Changes in Demand",ylab="Price", xlab="Quantity",xlim=c(0,5),ylim=c(0,5),bty='n')
curve(6-x,from=0,to=3.5,add=TRUE,col='blue')
curve(4-x,from=0,to=3.5,add=TRUE,col='blue')
text(x=3.6,y=2.5,expression("S"^1),col='blue')
text(x=3.6,y=1.5,expression("S"^0),col='black')
text(x=3.6,y=0.5,expression("S"^2),col='blue')
arrows(x0=2.6,y0=2.2,x1=2,y1=2.2,length=0.1,angle=30,col='blue',lwd=2)
arrows(x0=2.4,y0=2.8,x1=3,y1=2.8,length=0.1,angle=30,col='blue',lwd=2)
text(x=.9,y=2.2,'Decrease in demand')
text(x=4.1,y=2.8,'Increase in demand')
```

__change in quantity demanded__, When we graph the demand curve for good $X$, we hold everything but the price of $X$ constant. A representative demand curve is given by the center curve above. A movement along this curve is called a change in quantity demanded.

__change in quantity supplied__, A movement along the supply curve.

__change in supply__, Variables that affect the position of the supply curve are called _supply shifters_, and they include the prices of inputs, the level of technology, the number of firms in the market, taxes, and producer expectations. Whenever one or more of these variables change, the position of the entire supply curve shifts. Such a shift is known as a __change in supply__. The shift from $S^0$ to $S^2$ in the in the figure below represents a _decrease in supply_ since producers sell less of the product at each price.

```r
curve(x-0,from=0,to=3.5,main="Changes in Suppy",ylab="Price", xlab="Quantity",xlim=c(0,5),ylim=c(0,5),bty='n')
#axis(side=1,at=0:10,labels=FALSE)
#axis(side=2,labels=FALSE)
curve(x-1,from=0,to=3.5,add=TRUE,col='blue')
curve(x+1,from=0,to=3.5,add=TRUE,col='blue')

text(x=3.6,y=4.6,expression("S"^1),col='blue')
text(x=3.6,y=3.6,expression("S"^0),col='black')
text(x=3.6,y=2.6,expression("S"^2),col='blue')
arrows(x0=2.6,y0=2.3,x1=3.2,y1=2.3,length=0.1,angle=30,col='blue',lwd=2)
arrows(x0=2,y0=2.3,x1=1.5,y1=2.3,length=0.1,angle=30,col='blue',lwd=2)
text(x=.6,y=2.4,'Decrease in supply')
text(x=4.2,y=2.3,'Increase in supply')
```

__Comparative Statics Analysis__, the study of the movement from one market equilibrium to another based on changes in demand, or supply, or both.

__complements__, Goods which increase (decrease) in their demand as prices decrease (increase) in price.

__consumer expectations__, A demand shifter that changes the position of the demand curve for a product. If consumers suddenly expect the price of a good to increase significantly next year, the demand of the good will increase today. In effect, buying the good today is a substitute for buying the car next year.

__consumer surplus__, is the value consumers get from a good but don't have to pay for it. This is an important concept because it tells managers how much extra money consumers would be willing to pay for a given amount of a purchased product. The graph represents the demand function for a given product. If a product sells at the price of $3, in which the producer produces a quantity of 2, then the area above that price, under the demand curve is the consumer surplus.

```r
curve(5-x,from=0,to=5,main="Changes in Demand",ylab="Price", xlab="Quantity",bty='n',axes=FALSE, col='blue')
segments(x0=0,y0=c(4,3,2,1),x1=c(1,2,3,4),y1=c(4,3,2,1),lty=2)
segments(x0=c(4,3,2,1),y0=0,x1=c(4,3,2,1),y1=c(1,2,3,4),lty=2)
segments(x0=0,y0=3,x1=2,y1=3,col='blue', lwd=5)
segments(x0=0,y0=3.2,x1=1.8,y1=3.2,col='blue', lwd=5)
segments(x0=0,y0=3.4,x1=1.6,y1=3.4,col='blue', lwd=5)
segments(x0=0,y0=3.6,x1=1.4,y1=3.6,col='blue', lwd=5)
segments(x0=0,y0=4,x1=1,y1=4,col='blue', lwd=5)
segments(x0=0,y0=4.2,x1=.8,y1=4.2,col='blue', lwd=5)
segments(x0=0,y0=4.4,x1=.6,y1=4.4,col='blue', lwd=5)
segments(x0=0,y0=4.6,x1=.4,y1=4.6,col='blue', lwd=5)
points(x=2,y=3,lwd=5,col='red')
axis(side=1)
axis(side=2)
```

__decrease in demand__, A leftward shift in the demand curve.

__decrease in supply__, A leftward shift in the supply curve.

__demand__, The willingness of consumers to purchase a good or service, often represented as a quantity demanded at a price in the _demand curve_.

__demand curve__, the graphical representation of the quantity demanded, by consumers, for a good or service at a set price.

__demand shifters__, These are variables other than the price of a good (such as consumer income, advertising expenditures, etc...) that influence demand are known as _demand shifters_.

__equilibrium price__, The equilibrium price in a competitive market is determined by the interactions of all buyers and sellers in the market. The consepts of market supply and market demand make this notion of interaction more percise: The price of a good in a competitive market is determined by the interaction of market supply and market demand for the good.

```r
curve(5-x,from=.5,to=4.5,main="Market Equilibrium",ylab="Price", xlab="Quantity",xlim=c(0,5),ylim=c(0,5),bty='n')
curve(4-x,from=.5,to=4.5,add=TRUE,col="green")
curve(x-0,from=.5,to=4.5,add=TRUE,col='black')
curve(x+1,from=.5,to=4.5,add=TRUE,col='green')
```
