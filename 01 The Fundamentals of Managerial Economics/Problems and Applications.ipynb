11. You have recently learned that the company where you work is being sold for $\$\text{300,000}$. The company's income statement indicates current profits of $\$\text{11,000}$, which have yet to be paid out as dividends. Assuming the company will remain a "going concern" indefinitely and the interest rate will remain constant at 9 percent, at what constant rate does the owner believe that profits will grow? Does this seem reasonable? (LO5, LO7)

#### Answer
The formula for the present value of a firm at rate $i$ and growth rate $g$, is:
$$
\begin{align*}
\text{PV}_{Firm} = \pi_0\left(\frac{1 + i}{i - g}\right)
\end{align*}
$$
Below is a graph of the present value as a function of the growth rate. From this graph the owner expects the company's profits to grow 5 percent a year. This seems like a reasonable growth rate:

```r
curve( (11000*(1.09/(.09 - x))),
       from=.048,
       to=.052,
       type='l',
       main="Present Value as a function of growth",
       xlab="growth rate",
       ylab="present value")
segments(x0=0.01,y0=300000,x1=0.07,y1=300000,lty=1,col='blue')
segments(x0=0.05,y0=285000,x1=0.05,y1=305000,lty=1,col='red')
```

12. You are in the market for a new refrigerator for your company's lounge, and you have narrowed the search down to two models. The energy-efficient model sells for $\$\text{1,700}$ and will save you $\$45$ in electricity costs at the end of each of the next five years. The standard model has features similar to the energy-efficient model but provides no future saving in electricity costs. It is priced at only $\$\text{1,500}$. Assuming your opportunity cost of funds is 6 percent, which refrigerator should you purchase? (LO1, LO5, LO7)

#### Answer
The difference between the energy-efficient model and the standard model is $\$200$. If the present value of the $\$45$ savings for each of the next five years is greater than $\$200$, then the energey-efficient model should be purchased. If the present value of the savings is less than $\$200$, then the standard model should be purchased. Therefore, according to the following calculation, you should stick to the standard model.

```r
-round(pv.annuity(r=.06, n=5, pmt=45),digits=0)
```

13. You are the human resources manager for a famous retailer and are trying to convince the president of the company to change the structure of employee compensation. Currently, the company's retail sales staff is paid a flat hourly wage of $\$20$ per hour for each eight-hour shift worked. You propose a new pay structure whereby each salesperson in a store would be compensated $\$10$ per hour, plus $1$ percent of that store's daily profits. Assume that, when run efficiently, each store's maximum daily profits are $\$\text{25,000}$. Outline the arguments that support your proposed plan. (LO1, LO7) 

#### Answer
With a flat rate, there is no incentive for the salesperson to increase sales, and therefore profits, for the firm. The salesperson will make $\$160$ per day regardless if a sale is made during the day at all. The new compensation structure for the salespeople will now set a floor of sales of at least $\$80$ a day, per salesperson working, if the salespeople want to maintain their previous earnings. The new structure also incentives the salespeople to increase sales, up to the maximum daily profits of $\$\text{25,000}$. If the salesforce working that day can achieve that maximum daily profits, each salesperson working will make $\$330$ for the day.

14. Jamie is considering leaving her current job, which pays $\$\text{75,000}$ per year, to start a new company that develops applications for smartphones. Based on market research, she can sell about $\text{50,000}$ units during the first year at a price of $\$4$ per unit. With annual overhead costs and operating expenses amounting to $\$\text{145,000}$, Jamie expects a profit margin of $20$ percent. This margin is $5$ percent larger than that of her largest competitor, Apps, Inc. (LO2, LO3, LO7)

a. If Jamie decides to embark on her new venture, what will her accounting costs be during the first year of operation? Her implicit costs? Her opportunity costs?

The _accounting costs_ for the first year of operations will be the overhead costs and operating expenses that were calculated to be $\$\text{145,000}$. The _implicit cost_ is the loss of her current wages $\$\text{75,000}$. Therefore, the _opportunity cost_ is $\$\text{145,000} + \$\text{75,000} = \$\text{220,000}$.

b. Suppose that Jamie's estimated selling price is lower than originally projected during the first year. How much revenue would she need in order to earn positive accounting profits? Positive economic profits?

The _accounting profits_ of $\text{50,000} \times \$4 - \$\text{145,000} = \$\text{55,000}$ is already positive. However, the _economic profits_ which uses _opportunity cost_ instead of _accounting cost_ is $\text{50,000} \times \$4 - \$\text{145,000} - \$\text{75,000} = -\$\text{20,000}$. Therefore, in order to earn positive _economic profits_, selling $\text{50,000}$ units, the profits would need to be greater than the opportunity costs (accounting costs + implicit cost of losing her salary) which is $\$\text{220,000}$. The unit price would have to be $\gt$ $\frac{\$\text{220,000}}{\text{50,000}} = \$4.40$
